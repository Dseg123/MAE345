```python
from scipy.linalg import solve_continuous_are
import numpy as np

g = 9.81
A = np.array([[0, 1, 0, 0], [-2*g, 0, g, 0], [0, 0, 0, 1], [2*g, 0, -2*g, 0]])
B = np.array([[0, -1, 0, 2]]).T
Q = np.eye(4)
R = np.array([[1]])

S = solve_continuous_are(a = A, b = B, q = Q, r = R)
K = -np.linalg.inv(R) @ B.T @ S
print("S:", S)
print("K:", K)
```
